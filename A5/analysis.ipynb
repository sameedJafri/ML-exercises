From the quick analysis of the data we can see the data is simple with only 2 columns. No missing values.

```python
import pandas as pd
from datasets import load_dataset
```

```python
ds = load_dataset("stanfordnlp/imdb")

data = pd.DataFrame(ds["train"])
print("IMDB Dataset loaded successfully!")

print("Dataset Information:")
print(f"\tNumber of samples: {len(data)}")
print(f"\tShape of data: {data.shape}")
print(f"\tColumns: {data.columns.tolist()}")
print(f"\tUnique labels: {data['label'].unique()}\n")
print(f"Label distribution:\n{data['label'].value_counts()}\n")
print(f"Missing values:\n{data.isna().sum()}\n")
data.info()

print("\nSample data:")
print(data.head())

```
